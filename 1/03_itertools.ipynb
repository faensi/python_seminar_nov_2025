## 1 itertools Modul

Das `itertools` Modul bietet viele nützliche Funktionen für die Arbeit mit Iteratoren. <br>
Es ist Teil der Standardbibliothek und sehr effizient.


```python
import itertools

# cycle: Wiederholt einen Iterator unendlich
counter = itertools.cycle([1, 2, 3])
print([next(counter) for _ in range(7)])
```

## 2 chain

`chain()` verbindet mehrere Iterables zu einem einzigen Iterator.


```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
chained = list(itertools.chain(list1, list2))
print(chained)
```

## 3 combinations und permutations

- `combinations()`: Alle möglichen Kombinationen (ohne Wiederholung, Reihenfolge unwichtig) <br>
- `permutations()`: Alle möglichen Permutationen (ohne Wiederholung, Reihenfolge wichtig)


```python
items = ['A', 'B', 'C']
combs = list(itertools.combinations(items, 2))
perms = list(itertools.permutations(items, 2))
print("Combinations:", combs)
print("Permutations:", perms)
```

## 4 product

`product()` erzeugt das kartesische Produkt von mehreren Iterables.


```python
# Kartesisches Produkt
result = list(itertools.product([1, 2], ['a', 'b']))
print(result)
```
