## 1 Was ist Rekursion?

Rekursion ist eine Technik, bei der eine Funktion sich selbst aufruft. <br>
Jede rekursive Funktion braucht einen **Base Case** (Abbruchbedingung), um zu terminieren.


```python
# Beispiel: Fakultät
def factorial(n):
    # Base Case
    if n <= 1:
        return 1
    # Rekursiver Fall
    return n * factorial(n - 1)

print(factorial(5))
```

## 2 Fibonacci-Folge

Die Fibonacci-Folge ist ein klassisches Beispiel für Rekursion.


```python
def fibonacci(n):
    # Base Cases
    if n <= 0:
        return 0
    if n == 1:
        return 1
    # Rekursiver Fall
    return fibonacci(n - 1) + fibonacci(n - 2)

print([fibonacci(i) for i in range(10)])
```

## 3 Rekursion vs Iteration

Viele rekursive Probleme können auch iterativ gelöst werden.


```python
# Iterative Fakultät
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial_iterative(5))
```

#### Lösung:


```python
# Musterlösung (a)
def sum_list(lst):
    if not lst:
        return 0
    return lst[0] + sum_list(lst[1:])

print(sum_list([1, 2, 3, 4, 5]))

# Musterlösung (b)
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)

print(power(2, 5))

# Musterlösung (c)
def count_down(n):
    if n < 0:
        return
    print(n)
    count_down(n - 1)

count_down(5)
```
